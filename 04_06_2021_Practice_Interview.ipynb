```python
#if int is 1 return 1
#if divisible by 2 return half
#if not divisible by 2 return the int*3 + 1

def calculation(num):
    if num == 1:
        return num
    elif num % 2 == 0:
        return int(num/2)
    else:
        return int(num*3 + 1)
    
```

```python
calculation(5)
```

```python
#takes int as input continue to implement previous function until it gets to one
#return the number of iterations

def counter(num):
    counter = 0
    while num > 1:
        num = calculation(num)
        counter += 1
    return counter
```

```python
counter(5)
```

```python
sample = [1,2,3,4,[5,6],[7,8],9]
```

```python
def flatten_list(sample):
    new_list = []
    for elem in sample:
        if not isinstance(elem, list):
            new_list.append(elem)
        else:
            new_list.extend(elem)
    return new_list
```

```python
flatten_list(sample)
```

```python
#What if you have a nested list like [1,2,[3,4,[5,6,[7]]],8,9] turn it into a flattened list.

def flatten_nested_lst(lst):
    new_lst = []
    for elem in lst:
        if not isinstance(elem,list):
            new_lst.append(elem)
        else:
            new_lst.extend(flatten_nested_lst(elem))
    return new_lst
```

```python
flatten_nested_lst(lst)
```

```python
flatten_nested_lst(another_example)
```
